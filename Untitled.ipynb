```python
import http.client
import json

connection = http.client.HTTPConnection('api.football-data.org')
headers = { 'X-Auth-Token': '218888a9ec30477695209eb7fd05d26a' }
connection.request('GET', '/v2/competitions/2003/matches', None, headers )
response = json.loads(connection.getresponse().read().decode())

print (response)
```

```python
import pandas as pd
df = pd.json_normalize(response)
```

```python
df.head()
```

```python
datos_por_partido_json = df.iloc[0,1]
```

```python
dpp_df = pd.json_normalize(datos_por_partido_json)
```

```python
dpp_df.head(2)
```

```python
fecha_actual = dpp_df['season.currentMatchday'][0]
```

```python
dpp_df.drop(['stage','status','score.extraTime.homeTeam', 'score.extraTime.awayTeam', 'score.penalties.homeTeam', 'score.penalties.awayTeam','score.duration','season.currentMatchday','group', 'odds.msg', 'referees', 'lastUpdated', 'season.id', 'season.startDate', 'season.endDate'], axis=1, inplace=True)
```

```python
partidos_jugados = dpp_df.loc[dpp_df['matchday'] <= fecha_actual]
```

```python
partidos_jugados.tail()
```

```python
partidos_jugados['partidos_ganados'] = partidos_jugados.apply(lambda x: x['homeTeam.name'] if x['score.winner'] == 'HOME_TEAM' else ( x['awayTeam.name'] if x['score.winner'] == 'AWAY_TEAM' else 'asd'), axis=1)
partidos_jugados['partidos_empatados'] =  partidos_jugados.apply(lambda x: x['homeTeam.name'] if x['score.winner'] == 'DRAW' else 'asd', axis=1)
partidos_jugados['partidos_empatados_loc'] =  partidos_jugados.apply(lambda x: x['awayTeam.name'] if x['score.winner'] == 'DRAW' else 'asd', axis=1)
```

```python
partidos_jugados.head(2)
```

```python
agrup = partidos_jugados.groupby('homeTeam.name')
```
